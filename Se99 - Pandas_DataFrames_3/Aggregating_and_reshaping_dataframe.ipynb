### The .agg() method and Named Aggregations

#### You can name aggregated columns upon creation to avoid multi-index columns

```python
retail
```

```python
retail.groupby(["family","store_nbr"]).agg(["sum","mean"]) # it means the sum and mean operation will apply on remaining numeric columns.
```

```python
retail.groupby(["family","store_nbr"]).agg({"sales":"sum","onpromotion":"mean"})
```

```python
retail.groupby(["family","store_nbr"]).agg({"sales":["sum","mean"],"onpromotion":["min","max"]})
```

```python
(retail.groupby(
    ["family","store_nbr"]).agg(
    sales_sum=("sales","sum"),
    sales_mean=("sales","mean"),
    onpromotion_max=("onpromotion","max")
)
)
```

```python
retail2=retail
retail2
```

### .transform()  method  behaves as window functions in SQL

```python
retail2.assign(
    store_sales=retail2.groupby("store_nbr")["sales"].transform('sum').round())
```

```python
retail2.assign(
    store_sales=retail2.groupby(
        "store_nbr")["sales"].transform('sum').round(),
    sales_above_threshold_1000=lambda x: x["sales"]-1000)
```

### .pivot_table()  method 

#### let's you create Excel-style pivot tables

```python
retail
```

```python
retail.pivot_table(index="family",
                 columns="store_nbr",
                 values="sales",
                  aggfunc="sum").round()
```

```python
retail.pivot_table(index="family",
                 columns="store_nbr",
                 values="sales",
                  aggfunc="sum",
                  margins=True).round()
```

```python
retail.query(
    "family in ['AUTOMOTIVE','BEAUTY','POULTRY','SEAFOOD'] and \
             store_nbr in [1,2,3,4,5]").pivot_table(index="family",
                 columns="store_nbr",
                 values="sales",
                  aggfunc="sum",
                  margins=True).round()
```

### Multiple Aggregation functions in .pivot_table() method

```python
retail.query("family in ['AUTOMOTIVE','BEAUTY','SEAFOOD'] and \
             store_nbr in [1,2,3]").pivot_table(index="family",
                 columns="store_nbr",
                 values="sales",
                  aggfunc=("max","min"))
```

```python
retail.query("family in ['AUTOMOTIVE','BEAUTY','SEAFOOD'] and \
             store_nbr in [1,2,3]").pivot_table(index="family",
                 columns="store_nbr",
                  aggfunc=({"sales":["sum","mean"],"onpromotion":"max"}))
```

#### pivot_table()    vs groupby()

##### if the column argument isn't specified in a pivot table, it will return a table 
##### that's identical to one grouped by the index columns
