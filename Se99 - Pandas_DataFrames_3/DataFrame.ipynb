### Numpy select

```python
oil["buy"]=np.where(oil["price"]>100,"Too High","Buy")
oil
```

```python
oil['buy'].value_counts()
```

```python
conditions=[
    (oil["price"]>100),
    (oil["price"]<=100) & (oil["price"]>50),
    (oil["price"]<=50)
    
]

choices=["Don't Buy","Buy","Strong Buy"]

oil["buy2"]=np.select(conditions,choices,default='Missing')
oil
```

```python
oil["buy2"].value_counts(dropna=False)   # use dropna=False for see NaN value counts as well
```

```python
sample_df.assign(
on_promotion_flag=sample_df["onpromotion"]>0,
tax_amount=(sample_df["sales"]*0.05).round(2).map(lambda x: f"${x}")
)
```

```python
sample_df.assign(
on_promotion_flag=sample_df["onpromotion"]>0,
tax_amount=(sample_df["sales"]*0.05).round(2).map(lambda x: f"${x}"),
year=sample_df["date"].str[:4].astype("int")
)
```
